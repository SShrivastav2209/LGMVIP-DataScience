## *Image to Pencil Sketch with Python*

```python
from google.colab.patches import cv2_imshow
```

```python
import cv2
```

### *Reading the Image* 

```python
img = cv2.imread("/content/groot.jpeg")
cv2_imshow(img)
```

### *Converting it into a gray Image* 

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

### *Converting the gray Image into a negative Image* 

```python
inverted_image = cv2.bitwise_not(gray_image)
cv2_imshow(inverted_image)
```

### *Converting the inverted Image into a blurry Image* 

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
cv2_imshow(blurred)
```

### *Creating the pencil sketch by mixing the grayscale Image with the inverted blurry Image* 

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
cv2_imshow(pencil_sketch)
```
